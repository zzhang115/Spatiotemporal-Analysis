# Setting up our Schema

Spark can automatically create a schema for CSV files, but ours don't have headings. Let's set this up here:

```python
from pyspark.sql.types import StructType, StructField, FloatType, LongType, StringType
from pyspark.context import SparkContext
from pyspark.sql.session import SparkSession
from joblib import Parallel, delayed  
import multiprocessing

sc = SparkContext('local')
spark = SparkSession(sc)

feats = []
f = open('features.txt')
for line_num, line in enumerate(f):
    if line_num == 0:
        # Timestamp
        feats.append(StructField(line.strip(), LongType(), True))
    elif line_num == 1:
        # Geohash
        feats.append(StructField(line.strip(), StringType(), True))
    else:
        # Other features
        feats.append(StructField(line.strip(), FloatType(), True))
    
schema = StructType(feats)

print(schema)
```

```python
from pyspark.storagelevel import StorageLevel
```

# Creating a Dataframe

Let's load our CSV into a 'dataframe' - Spark's abstraction for working with tabular data (built on top of RDDs)

```python
#df = spark.read.format('csv').option('sep', '\t').schema(schema).load('/Volumes/evo/Datasets/NAM_2015_S/*')
df = spark.read.format('csv').option('sep', '\t').schema(schema).load('../nam/mini-sample.tdv')
# df = spark.read.format('csv').option('sep', '\t').schema(schema).load('../nam/nam_2015.tdv')

df.persist(StorageLevel.MEMORY_ONLY);
df.take(1)
df.count()
```

Where are you most likely to be struck by lightning? Use a precision of 4 Geohash characters and provide the top 3 locations.

# ML Predict Feature
The feature you predict? Answer: geopotential_height_lltw<br>
Features used to train the model Answer: geopotential_height_zerodegc_isotherm, temperature_surface, precipitable_water_entire_atmosphere, upward_long_wave_rad_flux_surface.<br>
How you partitioned your data? Answer: 90% / 10%<br>
Why the model choices make sense? Answer: At first I tried to use Random Regressor, but the predict result seems not good. For these features I picked from heatmap, the relationship of these feature seems linear regression, so I select the GeneralizedLinearRegression to train, and its final curve is relative match the diagonal.

```python
from pyspark.ml.feature import VectorAssembler

def prepare_data(dframe, predictors, target):
    assembler = VectorAssembler(inputCols=predictors, outputCol="features")
    output = assembler.transform(dframe)
    return output.select("features", target).withColumnRenamed(target, "label")


prepped = prepare_data(df,
    ["geopotential_height_zerodegc_isotherm", 
         "temperature_surface", 
         "precipitable_water_entire_atmosphere", 
         "upward_long_wave_rad_flux_surface"],
    "geopotential_height_lltw")

prepped.show()
(trainingData, testData) = prepped.randomSplit([0.9, 0.1])
```

```python
from pyspark.ml.regression import GeneralizedLinearRegression
from pyspark.ml.evaluation import RegressionEvaluator

rf = GeneralizedLinearRegression(family="gaussian", link="identity", maxIter=20, regParam=0.8)
model = rf.fit(trainingData)
predictions = model.transform(testData)

evaluator = RegressionEvaluator(
    labelCol="label", predictionCol="prediction", metricName="rmse")
rmse = evaluator.evaluate(predictions)
print("Root Mean Squared Error (RMSE) on test data = %g" % rmse)
```

```python
import matplotlib.pyplot as plt

p_df = predictions.select("label", "prediction").toPandas()

plt.suptitle('Generalized Linear Regression', fontsize=16)

minval = p_df[['label', 'prediction']].min().min()
maxval = p_df[['label', 'prediction']].max().max()
plt.axis([minval, maxval, minval, maxval])

plt.plot(p_df['label'], p_df['prediction'], '.', color='#2ba5f1')
plt.plot(range(int(minval), int(maxval)), range(int(minval), int(maxval)), lw=3, dashes=(10, 3), color='#000000', alpha=0.25, label='Ideal Predictor')
plt.show()
```
